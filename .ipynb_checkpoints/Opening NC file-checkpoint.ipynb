```python
import numpy as np
from numpy import *
import xarray as xr
import metpy
import pandas as pd
import matplotlib.pyplot as plt
```

```python
file_path_nc ="D:/Masters/Thesis/Data/Combined_FLDAS.nc"
#file_path_nc ="D:/Masters/Thesis/Data/ReGrid/Evap_tavg.nc"
data_array = xr.open_dataset(file_path_nc)
for variable in data_array.variables:
    print(variable)
which = 'Evap_tavg'
var = data_array.metpy.parse_cf(which)
values = var.isel(time=0)
lons = data_array.metpy.parse_cf('X')
lats = data_array.metpy.parse_cf('Y')
#if lats[0] > lats[-1]:
    #lats = lats[::-1]
    #values = values[:, ::-1]
unit = var.units
values
#missing_values_count = data_array.isnull().sum().item()

#print(f"Number of missing values in 'values': {missing_values_count}")
```

```python
df = pd.DataFrame(data=values.values)
df
```

```python
# Convert the data to a DataFrame
df = values.to_dataframe().reset_index()

# Pivot the DataFrame
pivot_df = df.pivot(index='lat', columns='lon', values=which)

# Display the pivoted DataFrame
pivot_df
# Specify the output file path
output_csv_path = "D:/Masters/Thesis/Data/ReGrid/Original/"+"original"+which+".csv"

# Export the pivoted DataFrame to a CSV file
pivot_df.to_csv(output_csv_path)

print(f"Pivoted DataFrame has been exported to {output_csv_path}")
```

```python

# Plot the data
import matplotlib.pyplot as plt
import cartopy.crs as ccrs
import cartopy.feature as cfeature
fig, ax = plt.subplots(figsize=(10, 5), subplot_kw={'projection': ccrs.PlateCarree()})
# Add coastlines and borders for context
ax.coastlines()
ax.add_feature(cfeature.BORDERS, linestyle=':')

# Plot the data
c = ax.pcolormesh(lons, lats, values, cmap='viridis', transform=ccrs.PlateCarree())
# Add a colorbar
fig.colorbar(c, ax=ax, orientation='vertical', label=unit)

# Add title and labels
ax.set_title(which)
ax.set_xlabel('Longitude')
ax.set_ylabel('Latitude')

plt.show()

```

## Multiple plots (for x and y)

```python
lst = xr.open_dataset('D:/Masters/Thesis/Data/Combined_FLDAS.nc')
display(lst)
time = lst['time']
#display(time)
# Assuming lst is an xarray Dataset
#lst = lst.transpose('time', 'x', 'y')  # Transpose if necessary
# Check if lat/lon are correctly labeled, otherwise rename them
lst = lst.rename({'y': 'y', 'x': 'x'})
for yr in range(2004,2005):
  filt_time = time.sel(time=time.dt.year==yr)
  display(filt_time)
  fig, axes = plt.subplots(nrows=3, ncols=4, figsize=(15, 10))
  axes = axes.flatten()
  i=1
  # Plot each month
  for t in filt_time:
      monthly_data = lst.sel(time=t)
      #print(t)
      # Assuming you want to plot the first variable in the dataset
      monthly_data['LST'].plot(ax=axes[i-1],cmap = 'viridis')
      #axes[i].set_title(f'Month {i}')
      i+=1

  # Adjust layout
  plt.subplots_adjust(left=0.1, right=0.9, top=0.9, bottom=0.1, hspace=0.4, wspace=0.4)
  #plt.savefig(f'/content/drive/MyDrive/Thesis_folder/pre_processed_data/IMERG/{yr}.png', bbox_inches='tight', dpi=300)
  plt.show()
countnan = lst['LST'].isnull().sum()
print(countnan)
```
