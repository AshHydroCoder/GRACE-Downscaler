## Regridding

```python
regrid_lat = np.arange(21.125,32.125,0.25)
regrid_lon = np.arange(76.125,92.125,0.25)
regrid = dataset.interp(y = regrid_lat,x=regrid_lon,method='nearest')
regrid
```

## Export to .tif file

```python
import numpy as np
import rasterio
from rasterio.transform import from_origin
data_array = regrid['band_data']
data_array = data_array.squeeze()

transform = from_origin(west=min(regrid_lon), north=max(regrid_lat), xsize=0.25, ysize=0.25)

# Export to GeoTIFF
with rasterio.open(
    'D:/Masters/Thesis/Data/regridded_dem.tif',  # Output file path
    'w',  # Write mode
    driver='GTiff',
    height=data_array.shape[0],  # Number of rows
    width=data_array.shape[1],   # Number of columns
    count=1,  # Number of bands (in this case, 1 band)
    dtype=data_array.dtype,  # Data type (e.g., float32)
    crs="EPSG:4326",  # Coordinate Reference System (WGS84)
    transform=transform  # Affine transformation
) as dst:
    dst.write(data_array.values, 1)  # Write data to the first band
```

```python
import rasterio
from rasterio.plot import show
import numpy as np
import matplotlib.pyplot as plt
from matplotlib import colors

# Function to calculate slope and aspect
def calculate_slope_aspect(dem, affine):
    # Calculate the gradients in the x and y directions
    dx, dy = np.gradient(dem, affine[0], affine[4])
    
    # Calculate slope
    slope = np.sqrt(dx**2 + dy**2)
    slope_degrees = np.degrees(np.arctan(slope))
    
    # Calculate aspect
    aspect = np.arctan2(-dy, dx)
    aspect_degrees = np.degrees(aspect)
    aspect_degrees = np.mod(aspect_degrees + 360, 360)  # Normalize to [0, 360]
    
    return slope_degrees, aspect_degrees

# Load and visualize the DEM .tif file
def visualize_dem_and_calculate_slope_aspect(dem_tif_path):
    with rasterio.open(dem_tif_path) as src:
        dem = src.read(1)  # Read the first band
        affine = src.transform

        # Visualize the DEM using plt.imshow for the colorbar
        plt.figure(figsize=(10, 8))
        plt.title('DEM')
        
        img = plt.imshow(dem, cmap='viridis', extent=(src.bounds.left, src.bounds.right, src.bounds.bottom, src.bounds.top),origin='lower')
        plt.colorbar(img, label='Elevation (m)')
        plt.show()
        
        # Calculate slope and aspect
        slope, aspect = calculate_slope_aspect(dem, affine)
        
        # Visualize slope
        plt.figure(figsize=(10, 8))
        plt.title('Slope (degrees)')
        plt.imshow(slope, cmap='RdYlBu', vmin=0, vmax=90, extent=(src.bounds.left, src.bounds.right, src.bounds.bottom, src.bounds.top), origin='lower')
        plt.colorbar(label='Slope (degrees)')
        plt.xlabel('Longitude')
        plt.ylabel('Latitude')
        plt.show()
        
        # Visualize aspect
        plt.figure(figsize=(10, 8))
        plt.title('Aspect (degrees)')
        plt.imshow(aspect, cmap='turbo', vmin=0, vmax=360, extent=(src.bounds.left, src.bounds.right, src.bounds.bottom, src.bounds.top), origin='lower')
        plt.colorbar(label='Aspect (degrees)')
        plt.xlabel('Longitude')
        plt.ylabel('Latitude')
        plt.show()

# Path to your DEM .tif file
dem_tif_path = 'D:/Masters/Thesis/Data/regridded_dem.tif'

# Run the visualization and calculation
visualize_dem_and_calculate_slope_aspect(dem_tif_path)
```

```python
import rasterio
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd

# Function to calculate slope and aspect
def calculate_slope_aspect(dem, affine):
    # Calculate the gradients in the x and y directions
    dx, dy = np.gradient(dem, affine[0], affine[4])
    
    # Calculate slope
    slope = np.sqrt(dx**2 + dy**2)
    slope_degrees = np.degrees(np.arctan(slope))
    
    # Calculate aspect
    aspect = np.arctan2(-dy, dx)
    aspect_degrees = np.degrees(aspect)
    aspect_degrees = np.mod(aspect_degrees + 360, 360)  # Normalize to [0, 360]
    
    return slope_degrees, aspect_degrees

# Load, calculate, and export DEM, slope, and aspect
def export_dem_slope_aspect_to_csv(dem_tif_path, output_csv):
    with rasterio.open(dem_tif_path) as src:
        dem = src.read(1)  # Read the first band
        affine = src.transform
        
        # Calculate slope and aspect
        slope, aspect = calculate_slope_aspect(dem, affine)
        
        # Get coordinates (longitude, latitude) for each DEM cell
        nrows, ncols = dem.shape
        x_coords, y_coords = np.meshgrid(np.arange(ncols), np.arange(nrows))
        
        # Calculate longitude and latitude using the affine transformation
        lon, lat = rasterio.transform.xy(affine, y_coords, x_coords, offset='center')
        
        # Flatten the arrays for export
        lon_flat = np.array(lon).flatten()
        lat_flat = np.array(lat).flatten()
        dem_flat = dem.flatten()
        slope_flat = slope.flatten()
        aspect_flat = aspect.flatten()
        
        # Create a pandas DataFrame
        df = pd.DataFrame({
            'Longitude': lon_flat,
            'Latitude': lat_flat,
            'DEM': dem_flat,
            'Slope (degrees)': slope_flat,
            'Aspect (degrees)': aspect_flat
        })
        
        # Export to CSV
        df.to_csv(output_csv, index=False)
        print(f"Data exported successfully to {output_csv}")

# Path to your DEM .tif file and output CSV file
dem_tif_path = 'D:/Masters/Thesis/Data/regridded_dem.tif'
output_csv = 'D:/Masters/Thesis/Data/DEM_slope_aspect.csv'

# Export the DEM, slope, and aspect to a CSV file
export_dem_slope_aspect_to_csv(dem_tif_path, output_csv)
```

# Repeat for different time 

```python
import pandas as pd

what = 'GRACE_Clipped_converted'
file_path_nc ="D:/Masters/Thesis/Data/ReGrid/"+what + ".nc"
G_dataset = xr.open_dataset(file_path_nc)
months_missing = G_dataset.attrs.get('months_missing', 'Attribute not found')
dates = months_missing.replace(",", ";").split(";")
missing_date = [(int(date.split("-")[0]), int(date.split("-")[1])) for date in dates]
missing_date = [pd.Timestamp(year=year, month=month, day=1) for year, month in missing_date]

#Step 1: Read the original CSV file
original_data = pd.read_csv('D:/Masters/Thesis/Data/DEM_slope_aspect.csv')

# Step 2: Create a date range from 2003-02-01 to 2020-12-31 (monthly frequency)
date_range = pd.date_range(start='2003-02-01', end='2020-12-31', freq='MS')

# Step 4: Convert 'missing_date' to datetime and filter out those dates from the date_range
missing_date = pd.to_datetime(missing_date, format='%Y-%m')
filtered_date_range = date_range[~date_range.isin(missing_date)]

# Step 5: Repeat the original data for each valid date (excluding the missing ones)
extended_data = pd.concat([original_data] * len(filtered_date_range), ignore_index=True)
extended_data['time'] = filtered_date_range.repeat(len(original_data)).values

# Step 6: Export the new data with the time dimension to a CSV file
extended_data.to_csv('D:/Masters/Thesis/Data/DEM_slope_aspect_new.csv', index=False)

print("New CSV file excluding the specified dates has been saved.")
```
